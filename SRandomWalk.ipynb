```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# P(X=1)=p
# n=number of runs
# m=number of random walks


def SRW(X0, p, n, m):
    probs = np.random.uniform(0,1,(n,m))
    RWprobs = pd.DataFrame(probs)
    bools = RWprobs > p
    RWIncrements = bools.replace([True, False], [1, -1])
    RWIncrements.loc[0:0] = X0
    RW = RWIncrements.cumsum()
    return RW
```

```python
Simulation1 = SRW(0, 0.5, 10**6, 50)
Simulation1SmootherPlot = Simulation1.iloc[::1000,:]

Simulation1SmootherPlot.plot(figsize=(20,10), linewidth=1, legend=False, title='Simulation')
```
